1. Horse-Specific Features
Past Performance: Previous race results, win/place/show history.
Speed Ratings: Historical speed ratings from past races.
Form: Current form or condition of the horse.
Age: Age of the horse.
Weight: Weight carried in the race.
Training: Training intensity and patterns leading up to the race.
Health Records: Veterinary reports, history of injuries or illnesses.
Breeding: Pedigree information, including sire and dam performance.
2. Jockey-Specific Features
Experience: Number of races ridden, win rates, and experience with specific horses.
Form: Current form or recent performance statistics.
Weight: Weight of the jockey.
3. Trainer-Specific Features
Reputation: Historical performance of horses trained.
Recent Form: Recent success rates of the trainer's horses.
Training Methods: Specific training methods or regimes known to the trainer.
4. Race-Specific Features
Race Distance: Distance of the race.
Race Class: Grade or class of the race.
Surface Type: Type of track surface (e.g., turf, dirt, synthetic).
Weather Conditions: Weather forecast and conditions on race day.
Track Condition: Track condition (e.g., firm, soft, muddy).
Race Number: Position or number of the race in the day's schedule.
5. Environmental Features
Location: The specific racecourse or venue.
Time of Year: Season or time of year, which can affect track conditions and horse performance.
Competition: Quality and performance of competing horses in the same race.

```python
#converted csv into dataframe
race_data=pd.DataFrame(race)
```

```python
horse_data=pd.DataFrame(horse)
```

```python
# data merging
horse_race=pd.merge(race_data,horse_data,on='rid')
```

```python
horse_race
```

```python
horse_race.info()
```

```python
horse_race.shape
```

```python
horse_race.isnull().sum()
```

```python

horse_race['currency'].fillna(horse_race['currency'].mode()[0],inplace=True)

horse_race['price'].fillna(horse_race['price'].mode()[0],inplace=True)
horse_race['overWeight'].fillna(horse_race['overWeight'].mode()[0],inplace=True)
horse_race['outHandicap'].fillna(horse_race['outHandicap'].mode()[0],inplace=True)
horse_race['headGear'].fillna(horse_race['headGear'].mode()[0],inplace=True)
horse_race['dist'].fillna(horse_race['dist'].mode()[0],inplace=True)
horse_race['positionL'].fillna(horse_race['positionL'].mode()[0],inplace=True)
horse_race['RPR'].fillna(horse_race['RPR'].mode()[0],inplace=True)
horse_race['TR'].fillna(horse_race['TR'].mode()[0],inplace=True)
horse_race['RPR'].fillna(horse_race['RPR'].mode()[0],inplace=True)
horse_race['positionL'].fillna(horse_race['positionL'].mode()[0],inplace=True)
horse_race['OR'].fillna(horse_race['OR'].mode()[0],inplace=True)
horse_race['hurdles'].fillna(horse_race['hurdles'].mode()[0],inplace=True)
```

```python
horse_race_cleaned=horse_race.dropna()
```

```python
horse_race_cleaned.isnull().sum()
```

```python
#seperation of categorical and numerical columns
import numpy as np
categorical_cols=horse_race_cleaned.select_dtypes(include=['object']).columns
numumerical_cols = horse_race_cleaned.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:",categorical_cols)
print("Numerical Variables:",numumerical_cols)
```

```python
#Plotting a pair plot for bivariate analysis
g = s.PairGrid(horse_race_cleaned,vars=['winningTime','res_win','overWeight','res_place','runners','saddle','position'])
#setting color
g.map_upper(s.scatterplot, color='crimson')
g.map_lower(s.scatterplot, color='limegreen')
g.map_diag(plt.hist, color='orange')
#show figure
plt.show()
```

```python
# Generate a heatmap to visualize the correlation matrix for the specified columns
column_name = ['winningTime', 'metric', 'overWeight', 'runners', 'res_win','res_place',
               'weight', 'margin']
s.heatmap(horse_race_cleaned[column_name].corr(), annot= True)
```
